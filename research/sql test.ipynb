```python
def create_table(pconn, df, table_name):

    cursor = pconn.cursor()

    # Define the table name and columns in your database
    columns = ', '.join([f"`{col}`" for col in df.columns])  # Wrap each column name in backticks

    # Prepare the SQL statement template for insertion
    placeholders = ', '.join(['%s'] * len(df.columns))
    sql = f"INSERT INTO {table_name} ({columns}) VALUES ({placeholders})"
    print(sql)
    df = df.fillna(-1)
    # Iterate over the rows in the DataFrame and insert them into the database
    for row in df.itertuples(index=False, name=None):
        cursor.execute(sql, row)
    
    print('Created table: ', table_name)
    # Commit the transaction
    pconn.commit()

    # Close the cursor
    cursor.close()
```
